```python
crime_datetime_format = np.array(pd.to_datetime(crime['Date']), dtype=np.datetime64)
crime_numeric_format = pd.to_numeric(crime_datetime_format)
median_numeric_format = np.median(crime_numeric_format)
median_datetime_format = pd.to_datetime(median_numeric_format)
median_datetime_format
```

```python
crime.Date.value_counts()
```

```python
crime.Date=pd.to_datetime(crime.Date)
crime.Date
```

```python
crime['Month']=crime.Date.dt.month
crime.head(1500)
```

```python
crime.Month.value_counts()
```

```python
crime['Weekday']=crime.Date.dt.weekday_name
crime.head(1500)
```
